```python
import numpy as np

np.random.seed(42)
matrix_5x5 = np.random.randint(0, 101, size=(5, 5))

print("\n Oluşturulan 5x5 Rastgele Matris:")
print(matrix_5x5)

print("\n Matris İstatistikleri:")

ortalama = np.mean(matrix_5x5)
print(f"• Ortalama: {ortalama:.2f}")

standart_sapma = np.std(matrix_5x5)
print(f"• Standart Sapma: {standart_sapma:.2f}")

varyans = np.var(matrix_5x5)
print(f"• Varyans: {varyans:.2f}")

en_buyuk = np.max(matrix_5x5)
print(f"• En büyük değer: {en_buyuk}")

en_kucuk = np.min(matrix_5x5)
print(f"• En küçük değer: {en_kucuk}")

kosegen_toplami = np.trace(matrix_5x5)
kosegen_elemanlari = np.diag(matrix_5x5)
print(f"• Köşegenindeki elemanların toplamı: {kosegen_toplami}")
print(f"• Köşegen elemanları: {kosegen_elemanlari}")
```

```python
import numpy as np

np.random.seed(42)
sinav_puanlari = np.random.normal(loc=75, scale=15, size=1000)

sinav_puanlari = np.clip(sinav_puanlari, 0, 100)

print(f" 1000 öğrencinin sınav puanları simüle edildi")
print(f"İlk 10 puan örneği: {sinav_puanlari[:10].round(1)}")

print(f"\n İstatistiksel Sonuçlar:")

ortalama = np.mean(sinav_puanlari)
print(f"• Ortalama: {ortalama:.2f}")

medyan = np.median(sinav_puanlari)
print(f"• Medyan: {medyan:.2f}")

standart_sapma = np.std(sinav_puanlari)
print(f"• Standart Sapma: {standart_sapma:.2f}")

print(f"\n Başarı Analizi:")

dusuk_puan_sayisi = np.sum(sinav_puanlari < 50)
print(f"• 50'den düşük alan öğrenci sayısı: {dusuk_puan_sayisi}")

dusuk_puan_orani = (dusuk_puan_sayisi / 1000) * 100
print(f"• Başarısızlık oranı: %{dusuk_puan_orani:.1f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Öğrenci': ['Ali', 'Ayşe', 'Mehmet', 'Zeynep', 'Ahmet'],
    'Yaş': [20, 21, 19, 22, 20],
    'Bölüm': ['Bilgisayar', 'Fizik', 'Kimya', 'Bilgisayar', 'Fizik'],
    'Matematik': [70, 60, 80, 90, 55],
    'Fizik': [65, 75, 70, 85, 60],
    'Kimya': [80, 85, 65, 95, 70]
}

df = pd.DataFrame(data)
print("\n Oluşturulan DataFrame:")
print(df)

print("\n Her ders için ortalama puan")
ders_ortalamalari = df[['Matematik', 'Fizik', 'Kimya']].mean()
print("\n Ders Ortalamaları:")
for ders, ortalama in ders_ortalamalari.items():
    print(f"• {ders}: {ortalama:.1f}")

print("\n En yüksek matematik notunu alan öğrenci")
en_yuksek_mat_idx = df['Matematik'].idxmax()
en_yuksek_mat_ogrenci = df.loc[en_yuksek_mat_idx]
print(f"\n En yüksek matematik notu:")
print(f"• {en_yuksek_mat_ogrenci['Öğrenci']} - {en_yuksek_mat_ogrenci['Matematik']} puan")

print("\n Her öğrencinin not ortalamasını hesaplayan yeni sütun")
df['Not_Ortalamasi'] = df[['Matematik', 'Fizik', 'Kimya']].mean(axis=1)
print("\n Not ortalaması eklenen DataFrame:")
print(df[['Öğrenci', 'Matematik', 'Fizik', 'Kimya', 'Not_Ortalamasi']].round(1))

print("\n Bölümlere göre gruplayarak ortalama başarılar")
bolum_ortalamalar = df.groupby('Bölüm')[['Matematik', 'Fizik', 'Kimya', 'Not_Ortalamasi']].mean()
print("\n Bölüm Bazında Ortalamalar:")
print(bolum_ortalamalar.round(1))

print("\n Ortalaması 70'in üzerinde olan öğrenciler")
basarili_ogrenciler = df[df['Not_Ortalamasi'] > 70]
print(f"\n Başarılı öğrenciler (ortalama > 70):")
print(basarili_ogrenciler[['Öğrenci', 'Bölüm', 'Not_Ortalamasi']].round(1))
print(f"Toplam başarılı öğrenci: {len(basarili_ogrenciler)}/{len(df)}")

plt.style.use('default')
plt.rcParams['figure.figsize'] = (12, 8)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
dersler = ['Matematik', 'Fizik', 'Kimya']
renkler = ['#FF6B6B', '#4ECDC4', '#45B7D1']

for i, ders in enumerate(dersler):
    axes[i].hist(df[ders], bins=4, alpha=0.7, color=renkler[i], 
                edgecolor='black', linewidth=1.5)
    axes[i].set_title(f'{ders} Notları Dağılımı', fontweight='bold', fontsize=12)
    axes[i].set_xlabel('Not Aralığı')
    axes[i].set_ylabel('Öğrenci Sayısı')
    axes[i].grid(True, alpha=0.3)
    
    ortalama = df[ders].mean()
    axes[i].axvline(ortalama, color='red', linestyle='--', linewidth=2,
                   label=f'Ort: {ortalama:.1f}')
    axes[i].legend()

plt.suptitle(' Her Dersin Dağılımı - Histogramlar', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

bolum_genel_ort = df.groupby('Bölüm')['Not_Ortalamasi'].mean()
bars1 = axes[0,0].bar(bolum_genel_ort.index, bolum_genel_ort.values,
                      color=['#FF6B6B', '#4ECDC4', '#96CEB4'], alpha=0.8)
axes[0,0].set_title('Bölümlere Göre Genel Ortalama', fontweight='bold')
axes[0,0].set_ylabel('Ortalama')
axes[0,0].set_ylim(0, 100)
for bar, val in zip(bars1, bolum_genel_ort.values):
    axes[0,0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2,
                   f'{val:.1f}', ha='center', fontweight='bold')

positions = [(0,1), (1,0), (1,1)]
for i, ders in enumerate(dersler):
    row, col = positions[i]
    ders_bolum_ort = df.groupby('Bölüm')[ders].mean()
    bars = axes[row,col].bar(ders_bolum_ort.index, ders_bolum_ort.values,
                            color=renkler[i], alpha=0.8)
    axes[row,col].set_title(f'{ders} - Bölüm Ortalamaları', fontweight='bold')
    axes[row,col].set_ylabel('Ortalama')
    axes[row,col].set_ylim(0, 100)
    
    for bar, val in zip(bars, ders_bolum_ort.values):
        axes[row,col].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2,
                          f'{val:.1f}', ha='center', fontweight='bold')

plt.suptitle(' Bölümlere Göre Ortalamalar - Bar Grafikleri', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
